### Overview

This notebook allows you to load data from JDBC databases using Spark SQL.

```
%sql
DROP Table default.configurations
```

```
driver = "com.mysql.jdbc.Driver"
url = "jdbc:mysql://avikcloud.mysql.database.azure.com:3306/drtest"
table = "configuration"
user = "avikcloud@avikcloud"
password = "PranavLaya@6"
```

```
remote_table = spark.read.format("jdbc")\
  .option("driver", driver)\
  .option("url", url)\
  .option("dbtable", table)\
  .option("serverTimezone", "UTC")\
  .option("user", user)\
  .option("password", "PranavLaya@6")\
  .load()
```

```
remote_table.createOrReplaceTempView("configuration")
```

```
remote_table.write.format("parquet").saveAsTable("configurations")
```

This table will persist across cluster restarts as well as allow various users across different notebooks to query this data. However, this will not connect back to the original database when doing so.
